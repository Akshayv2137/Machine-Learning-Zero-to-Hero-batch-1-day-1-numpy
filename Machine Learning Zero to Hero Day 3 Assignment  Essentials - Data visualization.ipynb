```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
## Bivariate Analysis
sns.scatterplot(x = df['Pregnancies'], y=df['Glucose'], hue=df['BloodPressure'])
```

```python
import sys

if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")
```

```python
sns.pairplot(df, hue='BloodPressure', size=3) ## useful for smaller number of features
plt.show()
```

```python
#### Histogram (Univariate analysis)

plt.hist(df['Pregnancies'], bins=30, rwidth= 0.8, color='red')
plt.show()
```

```python
## Box plot

sns.boxplot(df['Pregnancies'], orient='h', color='b')
```
